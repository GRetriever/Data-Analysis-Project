```python
import pandas as pd
import sklearn
from sklearn.preprocessing import LabelEncoder,StandardScaler,MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score,KFold

label_encoder = LabelEncoder()
standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
model = DecisionTreeClassifier(criterion='entropy',random_state=1234)
kf = KFold(n_splits=5,shuffle=True,random_state=1234)
```

```python
df = pd.read_csv('./data/mushroom_train.csv')

train_df = pd.read_csv("./data/mushroom_train.csv")
test_df = pd.read_csv("./data/mushroom_test.csv")
```

```python
train_df = train_df.drop(['cap-shape','cap-surface','gill-size','stalk-shape','stalk-surface-above-ring','stalk-surface-below-ring'],axis=1)
test_df = test_df.drop(['cap-shape','cap-surface','gill-size','stalk-shape','stalk-surface-above-ring','stalk-surface-below-ring'],axis=1)
```

```python
x_train = train_df.drop(['mushroom_id','class'], axis=1)
y_train = train_df['class']
x_test = test_df.drop(["mushroom_id"], axis=1)
```

```python
name_columns = ['cap-color','bruises','odor','gill-attachment','gill-spacing','gill-color','stalk-root','stalk-color-above-ring','stalk-color-below-ring','veil-type','veil-color','ring-number','ring-type','spore-print-color','population','habitat']
for name in name_columns:
    x_train[name] = label_encoder.fit_transform(x_train[name])
```
