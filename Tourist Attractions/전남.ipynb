```python
import os
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
names = os.listdir('./전라남도')
```

```python
names = [name[:-4] for name in names]
```

```python
df_전라남도 = pd.DataFrame()

for name in names:
    df = pd.read_csv(f'./전라남도/{name}.csv',encoding='ANSI')
    df_전라남도 = pd.concat([df_전라남도,df],ignore_index=True)
```

```python
df_전라남도
```

```python
df_전라남도.groupby('기초지자체')[['방문자 수']].sum().sort_values(by='방문자 수',ascending=False)
```
